```python
import pandas as pd
import pickle
```

```python
def read_file(mode='../pokemons'):
    with open(mode, 'rb') as fp:
        n_list = pickle.load(fp)
        return n_list
```

```python
lista_pokemons = read_file()
```

<p>Tempo com loop na lista_pokemons</p>

```python
%%timeit
for pokemon in lista_pokemons:
    print(pokemon)
```

```python
df = pd.DataFrame(lista_pokemons)
df.head()
```

```python
%%timeit
for i, linha in df.iterrows():
    display(i)
```
